```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from statsmodels.tsa.stattools import ccf

def interactive_cross_correlation(df, x_col, target_col, max_lag=365, alpha=0.05):
    x = df[x_col].dropna()
    y = df[target_col].dropna()

    # Ensure equal length
    min_len = min(len(x), len(y))
    x, y = x[:min_len], y[:min_len]

    # Compute cross-correlation
    ccf_values = ccf(x, y)[:max_lag]

    # Compute significance level (approximation)
    n = len(x)
    critical_value = 1.96 / np.sqrt(n)  # 95% confidence interval

    # Extract significant values
    significant_lags = np.where(np.abs(ccf_values) > critical_value)[0]
    significant_values = ccf_values[significant_lags]

    # Create interactive plot
    fig = go.Figure()

    # Add cross-correlation values
    fig.add_trace(go.Scatter(
        x=np.arange(max_lag),
        y=ccf_values,
        mode='lines+markers',
        name="Cross-Correlation",
        marker=dict(color='blue')
    ))

    # Add significance threshold
    fig.add_trace(go.Scatter(
        x=np.arange(max_lag),
        y=[critical_value] * max_lag,
        mode='lines',
        name="95% CI Upper",
        line=dict(color='red', dash='dash')
    ))

    fig.add_trace(go.Scatter(
        x=np.arange(max_lag),
        y=[-critical_value] * max_lag,
        mode='lines',
        name="95% CI Lower",
        line=dict(color='red', dash='dash')
    ))

    # Highlight significant points
    fig.add_trace(go.Scatter(
        x=significant_lags,
        y=significant_values,
        mode='markers',
        name="Significant Lags",
        marker=dict(color='green', size=8, symbol='star')
    ))

    # Layout adjustments
    fig.update_layout(
        title=f"Interactive Cross-Correlation between {x_col} and {target_col}",
        xaxis_title="Lag (days)",
        yaxis_title="Cross-Correlation",
        template="plotly_white"
    )

    fig.show()

    # Return significant lags and values
    return pd.DataFrame({"Lag": significant_lags, "Cross-Correlation": significant_values})

# Example usage
significant_results = interactive_cross_correlation(df, x_col='Badulla_precipitation_sum', target_col='cabbage')
print(significant_results)
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Function to create lagged features
def create_lagged_features(df, target_col, lags_dict):
    df_lagged = df.copy()

    for col, lags in lags_dict.items():
        for lag in lags:
            df_lagged[f"{col}_lag{lag}"] = df[col].shift(lag)

    # Drop rows with NaN values (from lagging)
    df_lagged.dropna(inplace=True)
    return df_lagged

# Define significant lags from your analysis
significant_lags = {
    'Badulla_precipitation_sum': range(22, 60),
    'Nuwara Eliya_precipitation_sum': range(110, 174),
    'Badulla_relative_humidity_2m_mean': range(49, 60),
    'Nuwara Eliya_relative_humidity_2m_mean': [0, 35],
    'Nuwara Eliya_temperature_2m_mean': range(0, 77),
    'Badulla_temperature_2m_mean': range(27, 130),
    'probability_neutral':range(27,67), 
    'probability_positive':range(0,140), 
    'probability_negative':range(0,365),
    'cabbage': range(0, 60) 
}

# Apply lag transformation
df_lagged = create_lagged_features(df, target_col="cabbage", lags_dict=significant_lags)

# Define features (excluding the original target column)
feature_cols = [col for col in df_lagged.columns if col != "cabbage"]

# 1 day ahead 
df_lagged['1_day_ahead_cabbage'] = df_lagged['cabbage'].shift(-1)
df_lagged = df_lagged.drop(['cabbage'], axis=1)
df_lagged = df_lagged.dropna()

# Train-test split (time-based)
train_size = int(len(df_lagged) * 0.8)  # 80% train, 20% test
train, test = df_lagged.iloc[:train_size], df_lagged.iloc[train_size:]

X_train, y_train = train[feature_cols], train["1_day_ahead_cabbage"]
X_test, y_test = test[feature_cols], test["1_day_ahead_cabbage"]

# Initialize and train XGBoost model
xgb_model = xgb.XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(X_train, y_train)

# Make predictions
y_pred = xgb_model.predict(X_test)

# Function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Calculate metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

# Print results
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"R² Score: {r2:.4f}")
print(f"MAPE: {mape:.2f}%")  # MAPE as a percentage

# Feature importance plot for XGBoost
plt.figure(figsize=(10, 6))
xgb.plot_importance(xgb_model, max_num_features=40, importance_type='weight', title='Top 20 Features Importance', height=0.8)
plt.show()
```

```python
# Most concise version - returns only significant predictor names
def find_granger_causal_predictors(data, target_col, predictors, max_lag=4, alpha=0.05):
    """
    Find predictors that Granger-cause the target variable (significant only)
    
    Returns: List of predictor names that show significant Granger causality
    """
    causal_predictors = []
    
    for predictor in predictors:
        test_data = data[[target_col, predictor]].dropna()
        
        if len(test_data) > max_lag:
            try:
                gc_test = grangercausalitytests(test_data, maxlag=max_lag, verbose=False)
                
                # Check if any lag shows significance
                significant_found = False
                for lag in range(1, max_lag + 1):
                    p_value = gc_test[lag][0]['ssr_chi2test'][1]
                    if p_value < alpha:
                        significant_found = True
                        break
                
                if significant_found:
                    causal_predictors.append(predictor)
                    
            except Exception as e:
                print(f"Error testing {predictor}: {e}")
                continue
    
    return causal_predictors
```

```python
import pandas as pd
from statsmodels.tsa.api import VAR

def simple_lag_selection(df, max_lags=10):
    """
    Simple VAR lag order selection
    """
    model = VAR(df)
    result = model.select_order(maxlags=max_lags)
    
    print("VAR Lag Order Selection Results:")
    print(f"AIC: {result.aic}")
    print(f"BIC: {result.bic}")
    print(f"HQIC: {result.hqic}")
    print(f"FPE: {result.fpe}")
    
    # Get the most frequently recommended lag
    recommendations = [result.aic, result.bic, result.hqic, result.fpe]
    recommended_lag = max(set(recommendations), key=recommendations.count)
    
    print(f"\nRecommended lag: {recommended_lag}")
    return recommended_lag

# Usage
# optimal_lag = simple_lag_selection(your_dataframe, max_lags=8)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error
import xgboost as xgb
from statsmodels.tsa.api import VAR
import optuna
from optuna.samplers import TPESampler

def create_lagged_features(df, lags):
    """
    Create lagged features for time series forecasting
    """
    df_lagged = df.copy()
    
    # Create lagged features for each variable
    for col in df.columns:
        for lag in range(1, lags + 1):
            df_lagged[f'{col}_lag_{lag}'] = df[col].shift(lag)
    
    return df_lagged.dropna()

def objective(trial, X_train, y_train, X_val, y_val):
    """
    Optuna objective function for hyperparameter optimization
    """
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 10),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 10),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_float('gamma', 0, 5),
        'random_state': 42
    }
    
    model = xgb.XGBRegressor(**params)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    
    return rmse

def build_xgboost_forecast_model_optuna(df, target_variable, lags=10, test_size=0.2, 
                                      n_trials=100, timeout=None, cv_folds=5):
    """
    Build XGBoost model for time series forecasting with Optuna hyperparameter optimization
    """
    
    # Create lagged features
    df_lagged = create_lagged_features(df, lags)
    
    # Prepare features and target
    feature_columns = [col for col in df_lagged.columns if col not in df.columns]
    X = df_lagged[feature_columns]
    y = df_lagged[target_variable]
    
    # Split data (maintaining time order)
    split_idx = int(len(X) * (1 - test_size))
    X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
    y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]
    
    # Further split training data for validation
    val_split_idx = int(len(X_train) * 0.8)
    X_train_opt, X_val_opt = X_train.iloc[:val_split_idx], X_train.iloc[val_split_idx:]
    y_train_opt, y_val_opt = y_train.iloc[:val_split_idx], y_train.iloc[val_split_idx:]
    
    # Optuna optimization
    study = optuna.create_study(
        direction='minimize',
        sampler=TPESampler(seed=42)
    )
    
    study.optimize(
        lambda trial: objective(trial, X_train_opt, y_train_opt, X_val_opt, y_val_opt),
        n_trials=n_trials,
        timeout=timeout
    )
    
    print(f"Best trial: {study.best_trial.number}")
    print(f"Best RMSE: {study.best_trial.value:.4f}")
    print("Best parameters:")
    for key, value in study.best_params.items():
        print(f"  {key}: {value}")
    
    # Train final model with best parameters on full training data
    best_params = study.best_params.copy()
    best_params['random_state'] = 42
    
    final_model = xgb.XGBRegressor(**best_params)
    final_model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = final_model.predict(X_test)
    
    # Calculate metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    
    # Feature importance
    feature_importance = pd.DataFrame({
        'feature': feature_columns,
        'importance': final_model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    return {
        'model': final_model,
        'study': study,
        'X_train': X_train,
        'X_test': X_test,
        'y_train': y_train,
        'y_test': y_test,
        'y_pred': y_pred,
        'metrics': {
            'mse': mse,
            'mae': mae,
            'rmse': rmse
        },
        'feature_importance': feature_importance,
        'feature_columns': feature_columns,
        'best_params': best_params
    }

    
# Example usage with your VAR results
if __name__ == "__main__":
    
    # Your VAR results showed AIC:10, BIC:2, HQIC:2, FPE:10
    # Since AIC and FPE recommend 10 lags, let's use 10 for richer features
    # optimal_lags = 10
    # target_variable = 'cabbage'  # Choose your target variable
    
    # print(f"Building XGBoost model with {optimal_lags} lags...")
    
    # # Build XGBoost model
    # results = build_xgboost_forecast_model_optuna(df, target_variable, lags=optimal_lags)
    
    # # Print results
    # print("\n=== Model Performance ===")
    # print(f"RMSE: {results['metrics']['rmse']:.4f}")
    # print(f"MAE: {results['metrics']['mae']:.4f}")
    # print(f"MSE: {results['metrics']['mse']:.4f}")
    
    # print("\n=== Top 10 Feature Importance ===")
    # print(results['feature_importance'].head(10))
    
    # # Compare with different lag choices
    # print("\n=== Comparing Different Lag Choices ===")
    for lags in [2, 5, 10]:  # Test BIC recommendation (2) vs AIC recommendation (10)
        try:
            comp_results = build_xgboost_forecast_model_optuna(df, target_variable, lags=lags, test_size=0.2)
            print(f"Lags {lags}: RMSE = {comp_results['metrics']['rmse']:.4f}")
        except Exception as e:
            print(f"Lags {lags}: Error - {e}")

# Additional function for multi-step forecasting
def multi_step_xgboost_forecast(df, target_variable, lags=10, steps=5):
    """
    Multi-step forecasting using XGBoost
    """
    # Create lagged features
    df_lagged = create_lagged_features(df, lags)
    
    # Prepare features and target
    feature_columns = [col for col in df_lagged.columns if col not in df.columns]
    X = df_lagged[feature_columns]
    y = df_lagged[target_variable]
    
    # Train final model on all data
    final_model = xgb.XGBRegressor(
        n_estimators=100,
        max_depth=6,
        learning_rate=0.1,
        random_state=42
    )
    final_model.fit(X, y)
    
    # Make future predictions
    last_observation = df.iloc[-1:].copy()
    future_predictions = []
    
    current_features = df.iloc[-lags:].values.flatten().reshape(1, -1)
    
    for step in range(steps):
        # Predict next value
        next_pred = final_model.predict(current_features)[0]
        future_predictions.append(next_pred)
        
        # Update features for next prediction (this is simplified)
        # In practice, you'd need a more sophisticated approach for multi-step
        # forecasting with multiple variables
    
    return future_predictions

# Hyperparameter tuning version
def tune_xgboost_forecast(df, target_variable, lags=10):
    """
    XGBoost with hyperparameter tuning
    """
    from sklearn.model_selection import GridSearchCV
    
    # Create lagged features
    df_lagged = create_lagged_features(df, lags)
    feature_columns = [col for col in df_lagged.columns if col not in df.columns]
    X = df_lagged[feature_columns]
    y = df_lagged[target_variable]
    
    # Split data
    split_idx = int(len(X) * 0.8)
    X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
    y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]
    
    # Parameter grid for tuning
```
