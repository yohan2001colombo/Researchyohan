```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('/content/drive/My Drive/Research UG/codes/DataSet/news_market.csv')
df2 = pd.read_csv('/content/drive/My Drive/Research UG/codes/DataSet/news_sentiment_mean.csv')
df3 = pd.read_csv('/content/drive/My Drive/Research UG/codes/DataSet/selected_weather.csv')

cabbages_df = pd.read_csv('/content/drive/My Drive/Research UG/codes/DataSet/new_cabbage.csv')
tomatoes_df = pd.read_csv('/content/drive/My Drive/Research UG/codes/DataSet/new_tomato.csv')
carrots_df = pd.read_csv('/content/drive/My Drive/Research UG/codes/DataSet/new_carrot.csv')
```

# Combine news, weather and vegetable prices

```python
# Convert date columns to datetime
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Date'] = pd.to_datetime(df2['Date'])

# --- Create a full date index based on df2 ---
all_dates = pd.DataFrame({'Date': df2['Date'].unique()})
all_dates = all_dates.sort_values('Date')

# --- Merge df1 with all dates ---
df1_full = all_dates.merge(df1, on='Date', how='left')

# --- Fill missing values with 0 ---
df1_full = df1_full.fillna(0)

# Now merge df1_full with df2 if needed
df_news = df1_full.merge(df2, on='Date', how='left')
```

```python
# Convert date columns
for df in [cabbages_df, df_news, df3]:
    df['Date'] = pd.to_datetime(df['Date'])

# Merge (side-by-side)
merged_df = (
    df_news
    .merge(cabbages_df, on='Date', how='left')
    .merge(df3, on='Date', how='left')
)
merged_df.to_csv('cabbage_full.csv', index=False)
```
